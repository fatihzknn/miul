```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")

df.groupby(["sex", "embark_town", "class"]).agg({
    "age":"mean",
    "survived": "mean",
    "sex" : "count"
})
```

```python
df.pivot_table("survived", "sex","embarked", aggfunc = "std") # first argument intersection , second row, third column
```

```python
df.pivot_table("survived", "sex",["embarked","class"])
```

```python
df["new_age"] = pd.cut(df["age"], [0,10,18,25,40,90])
df.head()
```

```python
df.pivot_table("survived","sex",["new_age","class"])
```

```python

df = sns.load_dataset("titanic")
df["age2"] = df["age"]*2
df["age3"] = df["age"]*5
```

```python
df[["age","age2","age3"]].apply(lambda x:x/10).head()
```

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x:x/10).head()
```

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x: (x - x.mean()) / x.std()).head() #normalization
```

```python
def standart_scaler(col_name):
    return (col_name - col_name.mean())/ col_name.std()
df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head() #normalization
```

```python
df.loc[:, df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head() #normalization
```
