```python
bins = [-1, 7, 30, 90, df['SaleCheckInDayDiff'].max()]
```

```python
df["new_SaleCheckInDayDiff"] = pd.cut(df["SaleCheckInDayDiff"],bins = bins, labels=["Last Minuters", "Potential Planners","Planners", "Early Bookers"])
```

```python
df.head()
```

```python
pivot_df = pd.pivot_table(df, values=["Price"], index = ["SaleCityName", "ConceptName", "new_SaleCheckInDayDiff"], aggfunc = ["mean","count"])
```

```python
pivot_df
```

```python
agg_df = pd.pivot_table(df, values="Price", 
                        index = ["SaleCityName", "ConceptName", "Seasons"], 
                        aggfunc = "mean").reset_index().sort_values(by= "Price", ascending=False)
```

```python
agg_df
```

```python
pivot_df.reset_index()
```

```python
agg_df["sales_level_based"] = (agg_df["SaleCityName"]+"_" + agg_df["ConceptName"]+ "_" +agg_df["Seasons"]).str.upper()
```
