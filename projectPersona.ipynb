```python
new_df = df.groupby(["COUNTRY","SOURCE","SEX","AGE"])[["PRICE"]].mean().sort_values("PRICE", ascending = False).reset_index()
```

```python
new_df
```

```python
df['AGE']=  df['AGE'].astype("category")
```

```python
df.dtypes
```

```python
new_df['AGE']=  new_df['AGE'].astype("category")
```

```python
new_df.dtypes
```

```python
bins = [0, 18, 23, 30, 40, 70]
labels = ['0_18', '19_23', '24_30', '31_40', '41_70']  # Kategori isimleri
```

```python
new_df['age_group'] = pd.cut(new_df['AGE'], bins=bins, labels = labels)
```

```python
new_df.head()
```

```python
new_df["age_group"] =  new_df["age_group"].astype("object")
```

```python
(new_df["COUNTRY"]+ "_" + new_df["SOURCE"]+ "_" + new_df["SEX"]+ "_" + new_df["age_group"]).str.upper()
```

```python
new_df["customers_level_based"] = (new_df["COUNTRY"]+ "_" + new_df["SOURCE"]+ "_" + new_df["SEX"]+ "_" + new_df["age_group"]).str.upper()
```
